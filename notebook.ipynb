```python
##CARGAR Y ACCEDER LA BASE DE DATOS DE LA FELICIDAD

import pandas as pd
felicidad= pd.read_csv('https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-08-07/felicidad.csv')
felicidad
```

```python
#VARIABLES DE LA BASE DE DATOS FELICIDAD
print(felicidad.info())
```

```python
#LIMPIAR DATOS NULL
felicidad_cleaned= felicidad.fillna(0)
felicidad_cleaned
```

```python
#FUNCIONES PARA ENCONTRAR PRIMER Y ÚLTIMO AÑO DEL REGISTRO DE ESTUDIO
primer_anio = felicidad['anio'].min()
print("El primer año de registro del estudio fue ", primer_anio)
ultimo_anio = felicidad['anio'].max()
print("El último y más reciente registro del estudio fue", ultimo_anio)
```

```python
#CALCULAR ÍNDICE DE FELICIDAD
 
#DEFINIR PONDERACIÓN
p_escalera_vida = 0.2
p_log_pib = 0.15
p_soporte_social = 0.1
p_expectativa_vida = 0.1
p_libertad = 0.1
p_generosidad = 0.05
p_percepcion_corrupcion = 0.1
p_afecto_positivo = 0.05
p_afecto_negativo = 0.05
p_confianza = 0.05
p_calidad_democracia = 0.05
#CALCULAR INDICE POR FILA
felicidad_cleaned['IndiceDeFelicidad'] = (
    p_escalera_vida * felicidad_cleaned['escalera_vida'] +
    p_log_pib * felicidad_cleaned['log_pib'] +
    p_soporte_social * felicidad_cleaned['soporte_social'] +
    p_expectativa_vida * felicidad_cleaned['expectativa_vida'] +
    p_libertad * felicidad_cleaned['libertad'] +
    p_generosidad * felicidad_cleaned['generosidad'] -
    p_percepcion_corrupcion * felicidad_cleaned['percepcion_corrupcion'] +
    p_afecto_positivo * felicidad_cleaned['afecto_positivo'] -
    p_afecto_negativo * felicidad_cleaned['afecto_negativo'] +
    p_confianza * felicidad_cleaned['confianza'] +
    p_calidad_democracia * felicidad_cleaned['calidad_democracia']
)
felicidad_cleaned
```

```python
#CALCULAR ÍNDICE PROMEDIO MUNDIAL PARA EL AÑO 2005 Y 2018 RESPECTIVAMENTE

felicidad_mundial_2005 = felicidad_cleaned[felicidad_cleaned['anio'] == 2005]['IndiceDeFelicidad'].mean()
felicidad_mundial_2018 = felicidad_cleaned[felicidad_cleaned['anio'] == 2018]['IndiceDeFelicidad'].mean()

print(f"El índice de felicidad promedio a nivel mundial para el año 2018 es: {felicidad_mundial_2018}")
print(f"El índice de felicidad promedio a nivel mundial para el año 2005 es: {felicidad_mundial_2005}")
```

```python
#ACCESO A LA INFORMACIÓN PARA LOS DOS PAÍSES EN EL PRIMER Y ÚLTIMO AÑO DEL ESTUDIO
#MEDIR PARA 2 PAÍSES: NEW ZEALAND

felicidad_nz_2006 = felicidad_cleaned[(felicidad_cleaned['pais'] == 'Nueva Zelanda') & (felicidad_cleaned['anio'] == 2006)]

print(felicidad_nz_2006[['pais', 'anio', 'IndiceDeFelicidad' ]])
                         
felicidad_nz_2018 = felicidad_cleaned[(felicidad_cleaned['pais'] == 'Nueva Zelanda') & (felicidad_cleaned['anio'] == 2018)]

print(felicidad_nz_2018[['pais', 'anio', 'IndiceDeFelicidad']])
```

```python
#MEDIR PARA 2 PAÍSES: ECUADOR

felicidad_ecuador_2006 = felicidad_cleaned[(felicidad_cleaned['pais'] == 'Ecuador') & (felicidad_cleaned['anio'] == 2006)]
felicidad_ecuador_2018 = felicidad_cleaned[(felicidad_cleaned['pais'] == 'Ecuador') & (felicidad_cleaned['anio'] == 2018)]

print(felicidad_ecuador_2018[['pais', 'anio', 'IndiceDeFelicidad']])
print(felicidad_ecuador_2006[['pais', 'anio', 'IndiceDeFelicidad']])
```
